### 1. Getting Started with NumPy 

```python
import numpy as np

print("NumPy version:", np.__version__)

arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)
```

### 2. Creating Arrays

```python
arr_0_9 = np.arange(0, 10)
zeros_5 = np.zeros(5, dtype=int)
ones_5 = np.ones(5, dtype=int)
arr_values = np.array([10, 20, 30, 40])

print("0 to 9:", arr_0_9)
print("Five zeros:", zeros_5)
print("Five ones:", ones_5)
print("[10,20,30,40]:", arr_values)
```

### 3. Array Properties

```python
arr_props = np.array([1, 2, 3, 4])

print("Size:", arr_props.size)
print("Data type:", arr_props.dtype)
print("Shape:", arr_props.shape)
```

### 4. Indexing & Slicing

```python
arr_idx = np.array([5, 10, 15, 20, 25])

print("First element:", arr_idx[0])
print("Last element:", arr_idx[-1])
print("Index 1 to 3:", arr_idx[1:4])  # elements at index 1,2,3

arr_idx[arr_idx == 10] = 100
print("After changing 10 to 100:", arr_idx)
```

### 5. Simple Array Operations

```python
arr_ops = np.array([1, 2, 3, 4, 5])

print("Add 10:", arr_ops + 10)
print("Multiply by 2:", arr_ops * 2)

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("a + b:", a + b)
print("a - b:", a - b)
```

### 6. Useful NumPy Functions

```python
arr_fun = np.array([3, 6, 9, 12])

print("Min:", np.min(arr_fun))
print("Max:", np.max(arr_fun))
print("Sum:", np.sum(arr_fun))
print("Mean:", np.mean(arr_fun))
```

### 7. Optional Practice

```python
arr_1_20 = np.arange(1, 21)

even_only = arr_1_20[arr_1_20 % 2 == 0]
print("Even numbers 1..20:", even_only)

capped = arr_1_20.copy()
capped[capped > 10] = 10
print("Replace values > 10 with 10:", capped)

count_gt_5 = (arr_1_20 > 5).sum()
print("Count elements > 5:", count_gt_5)
```
